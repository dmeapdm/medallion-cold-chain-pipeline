# Bronze layer overview

The Bronze layer is responsible for ingesting raw telemetry and sensor metadata from cloud storage without applying business transformations. Its purpose is to preserve source fidelity, provide traceability, and establish the entry point for the Medallion architecture.

## What happens here

* configure secure access to storage
* read raw telemetry JSON files
* read sensor metadata CSV files
* define output paths for downstream curated layers

## Why this matters

Keeping the raw layer separate makes the pipeline easier to audit, debug, and evolve. It also supports reproducibility because the downstream Silver and Gold logic can always be traced back to the original input structure.
